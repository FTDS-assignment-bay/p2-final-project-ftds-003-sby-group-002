## Target Variable Distribution

```python
# Let's see how Target feature is distributed 
plt.figure(figsize=(18,6))
plt.subplot(121)
df.TARGET.value_counts().plot(kind='pie', autopct = "%1.0f%%", colors=['skyblue','orange'])
plt.subplot(122)
ax = df.TARGET.value_counts().plot(kind='barh', color=['skyblue','orange'])
for i,j in enumerate(df["TARGET"].value_counts().values):
    ax.text(10,i,j)
plt.title('TARGET Variable Distribution')
plt.show()
```

- 92% of Target value is 0 (Lancar Bayar) sebanyak 282686
- 8% of Target value is 1 (Gagal Bayar) sebanyak 24825


## AMT INCOME TOTAL

```python
# Set ukuran plot
sns.set(rc={'figure.figsize':(10,10)})

# Hitung jumlah entri untuk setiap nilai di kolom OCCUPATION_TYPE
occupation_counts = df['OCCUPATION_TYPE'].value_counts()

# Filter data dengan OCCUPATION_TYPE yang termasuk dalam 10 teratas
df_top_10 = df[df['OCCUPATION_TYPE'].isin(occupation_counts.index)]

# Buat barplot dengan hue='TARGET'
bars = sns.barplot(x='AMT_INCOME_TOTAL', y='OCCUPATION_TYPE', hue='TARGET', data=df_top_10, estimator=lambda x: sum(x)/len(x), ci=None, palette='viridis', edgecolor='None')

# Buat nilai pada ujung bar dengan ukuran font yang lebih kecil
for bar in bars.patches:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():.2f}', ha='left', va='center', fontsize=8)

plt.show()
```

**Statement**

- Dari bar chart diatas menampilkan count tertinggi pada jenis pekerjaan `Manager`, dari semua count lancar bayar sebanyak 260714 dan yang gagal sebanyak 254634.
- Dan terdapat juga jenis pekerjaan dengan count amt income total dengan pekerjaan `Cleaning Staff` sebanyak 131073 untuk yang lancar bayar dan yang gagal bayar sebanyak 128128.
- Tapi dari bar chart diatas, pada jenis pekerjaan `IT Staff` terjadi perbedaan yang cukup banyak dengan lancar bayar sebanyak 216281.31 sedangkan konsumen yang dianggap gagal bayar sebanya 172720.

## HOUSETYPE_MODE

```python
# Hitung rata-rata AMT_GOOD_PRICE untuk setiap kategori HOUSETYPE_MODE
average_contract_price = df.groupby(['HOUSETYPE_MODE', 'TARGET'])['AMT_CREDIT'].mean().reset_index()

# Set ukuran plot
sns.set(rc={'figure.figsize':(10,5)})

# Buat barplot dengan hue='TARGET'
bars = sns.barplot(x='AMT_CREDIT', y='HOUSETYPE_MODE', hue='TARGET', data=average_contract_price, estimator=lambda x: sum(x)/len(x), ci=None, palette='viridis', edgecolor='None')

# Buat nilai pada ujung bar dengan ukuran font yang lebih kecil
for bar in bars.patches:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{bar.get_width():.2f}', ha='left', va='center', fontsize=8)

plt.show()
```

**Statement**

Dari bar chart diatas menampilkan bahwa konsumen yang lancar bayar dengan jumlah terbanyak berada pada tipe rumah block of flats sebanyak 665522 dan yang gagal bayar sebanyak 610015. Sedangkan untuk yang lancar bayar dari tipe rumah specifi housing yaitu berjumlah sebanyak 515164 dan yang gagal bayar berjumlah sebanyak 469917. Dari feature `HOUSETYPE_MODE` kami mencocokan dengan feature `AMT_GOODS_PRICE`, kolom ini digunakan untuk dapat memberikan nilai pada jenis rumah konsumen yang melakukan pinjaman. Terdapat tipe rumah yang unknown, dapat kami asumsikan bahwa rumah yang ditempati bukan milik pribadi.
